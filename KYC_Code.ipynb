```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import auth
from google.auth import default
import gspread
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
workspsheet= gc.open_by_url('https://docs.google.com/spreadsheets/d/1GteGZ5w7uJ7l3l3Ce5y1mEMeBGaPqj0vHTZSfXy9bvo/edit?gid=2120361731#gid=2120361731')
```

```python
raw_submit_KYC_sheet = workspsheet.worksheet('Raw Data KYC')
raw_data = raw_submit_KYC_sheet.get_all_values()
df= pd.DataFrame(raw_data)
df.rename(columns=df.iloc[0], inplace = True)
df.drop(df.index[0], inplace = True)
df['shopee_uid'] = df['shopee_uid'].apply(lambda x:x.replace(',',''))
df['kyc_request_id'] = df['kyc_request_id'].apply(lambda x:x.replace(',',''))
df['driver_id'] = df['driver_id'].apply(lambda x:x.replace(',',''))
df['applicant_id'] = df['applicant_id'].apply(lambda x:x.replace(',',''))
df.reset_index(drop = True, inplace = True)
```

```python
df["requested"] = pd.to_datetime(df["requested"], format='%Y-%m-%d %H:%M:%S')
df["updated"] = pd.to_datetime(df["updated"], format='%Y-%m-%d %H:%M:%S')
df["updated_date"] = pd.to_datetime(df["updated_date"], format='%Y-%m-%d')
```

```python
df.sort_values(by='requested', inplace = True)
df.sort_values(by='updated', inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
df_profile_picture = df[df['request_type'] == '2. profile_picture']
df_new_applicant = df[df['request_type'] == '1. new_applicant']
df_profile_picture.reset_index(drop = True, inplace = True)
df_new_applicant.reset_index(drop = True, inplace = True)
```

```python
new_applicant_uid = pd.unique(df_new_applicant['shopee_uid'])
df_new_applicant_uid = pd.DataFrame(new_applicant_uid)
df_new_applicant_uid.rename(columns={0: 'shopee_uid'},inplace = True)
df_new_applicant_uid.reset_index(drop = True, inplace = True)
new_picture_uid = pd.unique(df_profile_picture['shopee_uid'])
df_picture_uid = pd.DataFrame(new_picture_uid)
df_picture_uid.rename(columns={0: 'shopee_uid'},inplace = True)
df_picture_uid.reset_index(drop = True, inplace = True)
```

```python
from datetime import date
general_new_applicant_uid_list = []
general_new_applicant_resubmit_times_list = []
general_new_applicant_last_status_list = []
general_new_applicant_last_update_list = []
general_new_applicant_avg_reply_time_list = []
general_new_applicant_city_applied_list = []
general_new_applicant_city_tier_list = []
general_new_applicant_weeknum_list = []
general_new_applicant_month_list = []

for i in range(df_new_applicant_uid.shape[0]):
  df_new_applicant_search = df_new_applicant.loc[df_new_applicant['shopee_uid'] == df_new_applicant_uid.at[i,'shopee_uid']]
  df_new_applicant_search.reset_index(drop = True, inplace = True)
  general_new_applicant_resubmit_times_list.append(df_new_applicant_search.shape[0])
  general_new_applicant_uid_list.append(df_new_applicant_uid.at[i,'shopee_uid'])
  general_new_applicant_last_status_list.append(df_new_applicant_search['status'].iloc[-1])
  general_new_applicant_last_update_list.append(df_new_applicant_search['updated'].iloc[-1])
  general_new_applicant_city_applied_list.append(df_new_applicant_search['city_applied'].iloc[-1])
  general_new_applicant_city_tier_list.append(df_new_applicant_search['city_tier'].iloc[-1])
  general_new_applicant_weeknum_list.append(df_new_applicant_search['weeknum'].iloc[-1])
  general_new_applicant_month_list.append(df_new_applicant_search['Month'].iloc[-1])
  update_time_list = []
  request_time_list = []
  reply_time = 0
  for j in range(df_new_applicant_search.shape[0]):
      update_time_list.append(df_new_applicant_search.at[j,'updated'])
      request_time_list.append(df_new_applicant_search.at[j,'requested'])
  if df_new_applicant_search.shape[0] > 1:
    for a in range(len(request_time_list)-1):
      reply_time += (request_time_list[a+1] - update_time_list[a]).days
      #print('reply_time',reply_time)
    avg_reply_time = reply_time/(df_new_applicant_search.shape[0]-1)
  else:
    avg_reply_time = None
  general_new_applicant_avg_reply_time_list.append(avg_reply_time)


final_new_applicant_submit_KYC_dict = {'shopee_uid': general_new_applicant_uid_list, 'resubmit_times': general_new_applicant_resubmit_times_list,
              'last_status': general_new_applicant_last_status_list, 'last_update_time': general_new_applicant_last_update_list,
              'avg_reply_time': general_new_applicant_avg_reply_time_list,
              'city_applied': general_new_applicant_city_applied_list,'city_tier': general_new_applicant_city_tier_list,
              'weeknum':  general_new_applicant_weeknum_list, 'month': general_new_applicant_month_list}

df_new_submit_new_applicant_KYC = pd.DataFrame(final_new_applicant_submit_KYC_dict)
df_new_submit_new_applicant_KYC['request_type'] = '1. new_applicant'
```

```python
general_picture_uid_list = []
general_picture_resubmit_times_list = []
general_picture_last_status_list = []
general_picture_last_update_list = []
general_picture_avg_reply_time_list = []
general_picture_city_applied_list = []
general_picture_city_tier_list = []
general_picture_weeknum_list = []
general_picture_month_list = []

for i in range(df_picture_uid.shape[0]):
  df_picture_search = df_profile_picture.loc[df_profile_picture['shopee_uid'] == df_picture_uid.at[i,'shopee_uid']]
  df_picture_search.reset_index(drop = True, inplace = True)
  general_picture_resubmit_times_list.append(df_picture_search.shape[0])
  general_picture_uid_list.append(df_picture_uid.at[i,'shopee_uid'])
  general_picture_last_status_list.append(df_picture_search['status'].iloc[-1])
  general_picture_last_update_list.append(df_picture_search['updated'].iloc[-1])
  general_picture_city_applied_list.append(df_picture_search['city_applied'].iloc[-1])
  general_picture_city_tier_list.append(df_picture_search['city_tier'].iloc[-1])
  general_picture_weeknum_list.append(df_picture_search['weeknum'].iloc[-1])
  general_picture_month_list.append(df_picture_search['Month'].iloc[-1])
  update_time_list = []
  request_time_list = []
  reply_time = 0
  for j in range(df_picture_search.shape[0]):
      update_time_list.append(df_picture_search.at[j,'updated'])
      request_time_list.append(df_picture_search.at[j,'requested'])
  if df_picture_search.shape[0] > 1:
    for a in range(len(request_time_list)-1):
      reply_time += (request_time_list[a+1] - update_time_list[a]).days
      #print('reply_time',reply_time)
    avg_reply_time = reply_time/(df_picture_search.shape[0]-1)
  else:
    avg_reply_time = None
  general_picture_avg_reply_time_list.append(avg_reply_time)

final_picture_submit_KYC_dict = {'shopee_uid': general_picture_uid_list, 'resubmit_times': general_picture_resubmit_times_list,
              'last_status': general_picture_last_status_list, 'last_update_time': general_picture_last_update_list,
              'avg_reply_time': general_picture_avg_reply_time_list,
              'city_applied': general_picture_city_applied_list,'city_tier': general_picture_city_tier_list,
              'weeknum':  general_picture_weeknum_list, 'month': general_picture_month_list}

df_new_submit_picture_KYC = pd.DataFrame(final_picture_submit_KYC_dict)
df_new_submit_picture_KYC['request_type'] = '2. profile_picture'
```

```python
df_new_submit_KYC = pd.concat([df_new_submit_new_applicant_KYC,df_new_submit_picture_KYC])
df_new_submit_KYC
```
